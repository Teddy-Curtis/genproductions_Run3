```python
import glob 


def readFile(filename):
    with open(filename, 'r') as f:
        file = f.read()
    return file

def saveFile(file, filename):
    with open(filename, 'w') as f:
        f.write(file)

new_run_card_file = "_run_card_13p6TeV.dat"
new_run_card = readFile(new_run_card_file)
print(new_run_card)
```

```python
all_run_cards = glob.glob(f"cards/mH*/idm_dilepton*/*run_card.dat")
```

```python
for run_card_loc in all_run_cards:
    # overwrite this run card with the new one 
    saveFile(new_run_card, run_card_loc)
```
